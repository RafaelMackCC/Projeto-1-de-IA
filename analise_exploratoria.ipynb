```python
import pandas as pd

# Convert 'Date' column to datetime objects
train_data['Date'] = pd.to_datetime(train_data['Date'])

# Extract numerical features from the date
# For example, you can extract week of year
train_data['WeekOfYear'] = train_data['Date'].dt.isocalendar().week

# Now you can calculate correlations, excluding the original 'Date' column
print("\nCorrelação entre as variáveis numéricas do dataset de treino")
print(train_data.drop(columns=['Date']).corr()) # Exclude 'Date' column from correlation
```

```python
# Explorar as correlações entre as variáveis
print("\nCorrelação entre as variáveis numéricas do dataset de treino")
print(train_data.corr())
```

```python
sns.histplot(train_data['Weekly_Sales'], bins=50, kde=True)
plt.title('Distribuição de Vendas Semanais')
plt.show()

# Visualizar correlação entre as variáveis
plt.figure(figsize=(10,8))
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de Calor das Correlações')
plt.show()
```
